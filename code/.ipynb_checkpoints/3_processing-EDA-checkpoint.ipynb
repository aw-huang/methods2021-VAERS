```julia
# install packages
import Pkg;
Pkg.add("DataFrames");
Pkg.add("CSV");
Pkg.add("Plots");
Pkg.add("StatsPlots");
Pkg.add("StatsBase");
Pkg.add("Missings");
Pkg.add("Dates")
```

```julia
# setup filepaths
path_source = string(@__DIR__,"\\..\\source");
path_dev = string(@__DIR__,"\\..\\dev");
```

```julia
# LOAD IN PACKAGES AND DATA
using DataFrames, CSV, Plots, StatsPlots, StatsBase, Missings
df = DataFrame(CSV.read(joinpath(path_dev,"19-21VAERSCOMB.csv"), DataFrame));
```

```julia
#EXPLORATORY DATA ANALYSIS
summary(df) #size of the dataset
```

```julia
show(names(df)) #display colnames
```

```julia
df = df[(.!ismissing.(df.AGE_YRS)) .& (df.AGE_YRS .>= 16), :] #filter age >= 16
first(df, 10)
```

```julia
vax_ct = countmap(df[!, :VAX_NAME])
sort(vax_ct, byvalue=true, rev=true)
death_ct = countmap(df[!, :DIED])
sort(death_ct, byvalue=true, rev=true)
threat_ct = countmap(df[!, :L_THREAT])
sort(threat_ct, byvalue=true, rev=true)
hos_ct = countmap(df[!, :HOSPITAL])
sort(hos_ct, byvalue=true, rev=true)
```

```julia
# Convert RECVDATE from String into DateTime
using Dates
df.RECVDATE = Date.(df.RECVDATE, "mm/dd/yyyy")
first(df, 10)
```

Initial Visualizations

```julia
#Days in hospital Histogram
histogram(filter(x -> !(ismissing(x) || isnan(x)), df[!, :HOSPDAYS]),
    bins=600,xlim=(0,100),xlabel="HOSPDAYS",labels="Frequency")
```

```julia
# Distribution over time
df_time = combine(groupby(combine(groupby(df, :VAERS_ID), :RECVDATE=>:RECVDATE), :RECVDATE), nrow => :COUNT)
first(df_time, 10)
```

```julia
plot(df_time[!,:RECVDATE], df_time[!,:COUNT],
    xlabel="Date", ylabel="Frequency of VAERS reports", title="Distribution of adult VAERS reports over time", legend=false)
# xticks=[df_time[!,:RECVDATE][i] for i in 1:5:length(df_time[!,:RECVDATE])]
```

```julia
# Distribution of Age Histogram
df_age = combine(groupby(df, :VAERS_ID), :AGE_YRS=>:AGE_YRS)
histogram(df_age[!, :AGE_YRS],
    xlabel="Age (years)", ylabel="Frequency of VAERS reports", legend=false)
```

```julia
df_sex = combine(groupby(combine(groupby(df, :VAERS_ID), :SEX=>:SEX), :SEX), nrow => :COUNT);
df_sex
```

Labeling reports as serious or non-serious

```julia
#labeling of events as serious or non-serious
a = (df.DIED .== "Y") .| (df.L_THREAT .== "Y") .| (df.HOSPITAL .== "Y") .| (df.X_STAY .== "Y") .| (df.DISABLE .== "Y")
df.SERIOUS_EVENT = replace(a, missing => 0)
show(names(df))
```

```julia
#count # of serious event
sum(df.SERIOUS_EVENT)
#size of entire dataset
summary(df) #size of the dataset
```

```julia
#SERIOUS_EVENT =1 is serious and SERIOUS_EVENT=0 is non-serious adverse event
#find most frequent serious adverse event - pull from SYMPTOM1, SYMPTOM2, SYMPTOM3, SYMPTOM4, SYMPTOM5
#find most frequent non-serious adverse event - pull from SYMPTOM1, SYMPTOM2, SYMPTOM3, SYMPTOM4, SYMPTOM5
df_nonserious = df[df.SERIOUS_EVENT .== 0, :];
df_serious = df[df.SERIOUS_EVENT .== 1, :];
#select columns - SERIOUS_EVENT, VAERS_ID, SYMPTOM1, SYMPTOM2, SYMPTOM3, SYMPTOM4, SYMPTOM5
```

Visualization for SERIOUS_EVENT

```julia
df_SERIOUS = combine(groupby(combine(groupby(df, :VAERS_ID), :SERIOUS_EVENT=>:SERIOUS_EVENT), :SERIOUS_EVENT), nrow => :COUNT);
df_SERIOUS
```

```julia
bar(df_SERIOUS[!, :COUNT],
    xlabel="Reported Serious Event", ylabel="Frequency of VAERS reports", legend=false, 
    xticks=(1:2, ["Not Serious", "Serious"]), formatter=:plain)
```

```julia
# Pie chart
pie(["Not Serious", "Serious"], 
    df_SERIOUS[!,:COUNT]/sum(df_SERIOUS[!,:COUNT]), 
    title = "Proportion of Serious vs. Not Serious VAERS reports", l = 1, legend=true, formatter=:plain)
```
