```python
import pandas as pd
import os
dirname = os.getcwd()
filename = os.path.join(dirname, '..\\dev\\19-21VAERSSYMPTOMS_dev.csv')
```

```python
df = pd.read_csv(filename)
```

```python
df.head(20)
```

```python
df.SYMPTOMS
```

```python
agg_df = df.groupby(['VAERS_ID']).agg({'SYMPTOMS': lambda s: set(s)})
```

```python
agg_df.head(20)
```

```python
filename = os.path.join(dirname, '..\\dev\\19-21VAERSSYMPTOMS_dev2.csv')
agg_df.to_csv(filename)
```
